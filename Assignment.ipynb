## Plotting

```python
dataForPlot = pd.read_csv(covidVaccine,na_values='')
dataForPlot.fillna(0, inplace = True)
dataForPlot
```

```python
firstDose = list(dataForPlot.iloc[:,0])
secondDose = list(dataForPlot.iloc[:,1])
date = range(len(dataForPlot.iloc[:,2]))

```

```python
plt.fill_between(date,firstDose,0,color = 'blue',alpha = 0.5,label = 'First Dose')
plt.fill_between(date,secondDose,0,color = 'red',alpha = 0.5, label = 'Second Dose')
plt.legend()
plt.xlabel('Days since start of Covid')
plt.ylabel('Number of Doses Administered e8')
```
